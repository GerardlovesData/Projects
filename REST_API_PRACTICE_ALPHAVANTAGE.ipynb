```python
# create list of stocks that can be used in for loop to download multiple stocks' information

stocklist=['AXP','AMGN','AAPL','BA','CAT','CSCO','CVX','GS','HD','HON','IBM','INTC','JNJ','KO','JPM','MCD','MMM','MRK','MSFT','NKE','PG','TRV','UNH','CRM','VZ','V','WBA','WMT','DIS','DOW']

#stocklist=['AXP','AMGN','AAPL','BA','CAT']

#stocklist=['AXP']

for stock in stocklist:
    print(stock)
```

```python
# Create list that can be converted to dataframe that can be exported to .csv
# add time.sleep(13).  13 second delay per call.  Alpha Vantage only allows 5 calls per minute

pricelist=[]

for stock in stocklist:
    url_loop = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol='+stock+'&apikey=6SMI3OGMY7ORQGEE'
    response_loop = requests.get(url_loop)
    response_json_loop=json.loads(response_loop.text)
    time.sleep(13)
    for date in response_json_loop['Monthly Time Series']:
        stocksymbol=response_json_loop['Meta Data']['2. Symbol']
        closeprice=response_json_loop['Monthly Time Series'][date]['4. close']
        pricelist.append([stocksymbol,date,closeprice])
        
        
        # write output from loop to csv
        # https://stackoverflow.com/questions/54900315/write-output-from-for-loop-to-a-csv-in-python
        
        
print(pricelist)
```

```python
# write output from loop to csv
# https://stackoverflow.com/questions/54900315/write-output-from-for-loop-to-a-csv-in-python

df = pd.DataFrame(pricelist, columns=['stock_ticker', 'date', 'month_end_price'])

print(df)

df.to_csv(r'''C:\Users\xxxxx\REST_API_PRACTICE_ALPHAVANTAGE_PRICELIST.csv''', sep=',', index=False)

df.to_excel(r'C:\Users\xxxxx\REST_API_PRACTICE_ALPHAVANTAGE_PRICELIST.xlsx', sheet_name='Pricelist', index=False)



```

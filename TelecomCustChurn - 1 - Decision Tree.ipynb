```python
from sklearn import tree

from sklearn.model_selection import train_test_split

(trainfeat, testfeat, trainchurn, testchurn)=train_test_split(features, churn, train_size=0.75, random_state=0)
```

```python
print(testfeat.iloc[:, 0].size, '\n')

print(trainfeat.iloc[:, 0].size)
```

```python
clf=tree.DecisionTreeClassifier(random_state=100, max_depth=3)
clf=clf.fit(trainfeat, trainchurn)

print(clf)
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
import pydotplus

dot_data = StringIO()  
tree.export_graphviz(clf, out_file=dot_data,  
                         feature_names=featlab)  
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())  
```

```python
# Function Name: np.logical_and()     pd.dataframe
# Importance of this function: Row Access:  Conditions to subset the data even more
# Returns: a more particular data frame
# Requirements: condition parameters
# Syntax:
# Comments: the example code below arrives at a numpy array that has Age, Income, Gender.  Income has be to <50K and >20K

# Obtain Age, Income, and Gender subset and turn it into an array
slice=np.array(data.loc[:, ['Age', 'Income', 'Gender']])

# slice[:, 1] in the code below refers to all rows of the Income column in slice
slice=slice[np.logical_and(slice[:, 1]<50000, slice[:, 1]>20000)]

# Convert array back to a dataframe
data2=pd.DataFrame(data=slice,columns=['Age', 'Income', 'Gender'])

data2






```

```python
# Function Name: np.logical_or()     pd.dataframe
# Importance of this function: Row Access:  Conditions to subset the data even more
# Returns: a more particular data frame
# Requirements: condition parameters
# Syntax:
# Comments: the example code below arrives at a numpy array that has Age, Income, Gender.  Age has to be 57 or Gender is Male

# Obtain Age, Income, and Gender subset and turn it into an array
slice2=np.array(data.loc[:, ['Age', 'Income', 'Gender']])

# slice[:, 1] in the code below refers to all rows of the Income column in slice. MUST USE 2 EQUAL SIGNS! ==
slice2=slice2[np.logical_or(slice2[:, 0]==57, slice2[:, 2]=='Male', slice2[:, 2]=='M')]

# Convert array back to a dataframe
data3=pd.DataFrame(data=slice2,columns=['Age', 'Income', 'Gender'])

data3
```

```python
# Function Name: np.logical_not()     pd.dataframe
# Importance of this function: Row Access:  Conditions to subset the data even more
# Returns: a more particular data frame
# Requirements: condition parameters
# Syntax:
# Comments: the example code below arrives at a numpy array that has Age, Income, Gender.  Age cannot be >21

# Obtain Age, Income, and Gender subset and turn it into an array
slice2=np.array(data.loc[:, ['Age', 'Income', 'Gender']])

# slice[:, 1] in the code below refers to all rows of the Income column in slice. MUST USE 2 EQUAL SIGNS! ==
slice2=slice2[np.logical_not(slice2[:, 0]>21)]

# Convert array back to a dataframe
data3=pd.DataFrame(data=slice2,columns=['Age', 'Income', 'Gender'])

data3
```

```python
# Function Name: iterrows() A generator that iterates over the rows of the frame.
#                Counter counts the frequency of each unique member of the derived column
# Importance of this function: Create derived columns based on conditions
# Returns:
# Requirements:
# Syntax:
# Comments: 
#           In other words, index below refers to row number.  
#           Row pertains to all contents of that row number (all columns)

from collections import Counter

for row_number, column in data.iterrows():
    if column["Age"]<=40 and column["Country"]=="United States":  # Dont forget to use double equal signs!
        data.loc[row_number, "Age_cat"]="US: Under or equal to 40" # This line creates the derived column
    else:
        data.loc[row_number, "Age_cat"]="Other"
        
print(Counter(data["Age_cat"]), "\n")

print(data.loc[:, "Age_cat"])
    
    


    
```

```python
# Function Name: groupby.mean()
# Importance of this function: aggregate a column of data and return a mean
# Returns:
# Requirements:
# Syntax: If you want to get the means of all of the other columns remove ['Age'].  pd.DataFrame converts the Series to a 
#         DataFrame
# Comments: the code below groups unique values of the Gender column and returns 
#           the mean age of each unique Gender value

pd.DataFrame(data.groupby('state').mean()['Age'])
```

```python
# Function Name: iterrows() A generator that iterates over the rows of the frame.
#                Counter counts the frequency of each unique member of the derived column
# Importance of this function: Create derived columns based on conditions
# Returns:
# Requirements:
# Syntax:
# Comments:

# Obtain a cleaner Gender column

for index, row in data.iterrows():
    if row["Gender"]=="Female" or row["Gender"]=="female" or row["Gender"]=="F" or row["Gender"]=="f":  # Dont forget to use double equal signs!
        data.loc[index, "Gender_clean"]="Female" # This line creates the derived column
    
    elif row["Gender"]=="Male" or row["Gender"]=="male" or row["Gender"]=="M" or row["Gender"]=="m":  # Dont forget to use double equal signs!
        data.loc[index, "Gender_clean"]="Male" # This line creates the derived column

    else:
        data.loc[index, "Gender_clean"]="Misc Gender Category"
        
data.loc[:, ["Gender", "Gender_clean", "Income"]]
```

```python
# Function Name: Imputing missing data by using median and fillna then create new DataFrame
# Importance of this function:  replace NaN and other values
# Returns:
# Requirements: from numpy import *
# Syntax:
# Comments: https://www.youtube.com/watch?v=gJRhXp8Sm4Y (minute 4:41)

import pandas as pd

from numpy import *

data5=pd.read_excel(r'''C:\Springboard\0 - Mental Health Raw Data.xlsx''', 
                   sheetname='Mental Health Data', 
                   usecols=['Timestamp', 'Age_Impute', 'Income', 'Gender', 'Country', 'state', 'self_employed', 'treatment'], 
                   index_col='Timestamp')

data5=data5[:100]

data5['Age_Impute_new']=data5['Age_Impute'].fillna(data5['Age_Impute'].median())


#data5['Age_Impute']

old_vs_new=pd.DataFrame(
    {'old': data5['Age_Impute'],
    'new': data5['Age_Impute_new']})

old_vs_new
```
